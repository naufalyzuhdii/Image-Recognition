```python
import cv2 
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
%matplotlib inline
```

Load & Plot Image

```python
image_1 = pltimg.imread('curr1.jpg')  
image_2 = pltimg.imread('curr2.jpg') 

plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.title("Original Image 1")
plt.imshow(image_1)

plt.subplot(1,2,2)
plt.title("Original Image 2")
plt.imshow(image_2)
plt.show()
```

Ubah jadi gray

```python


gray_image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY)
gray_image_2 = cv2.cvtColor(image_2, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.title("Original Image 1")
plt.imshow(gray_image_1, cmap = 'gray')

plt.subplot(1,2,2)
plt.title("Original Image 2")
plt.imshow(gray_image_2, cmap = 'gray')
plt.show()
```

SIFT

```python
sift = cv2.xfeatures2d.SIFT_create()

keypoints_1, descriptors_1 = sift.detectAndCompute(gray_image_1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(gray_image_2,None)

len(keypoints_1), len(keypoints_2)
```

Feature Matching BF

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

image_3 = cv2.drawMatches(gray_image_1, keypoints_1, gray_image_2, keypoints_2, matches[:50], gray_image_2, flags=2)
plt.imshow(image_3),plt.show()
```
